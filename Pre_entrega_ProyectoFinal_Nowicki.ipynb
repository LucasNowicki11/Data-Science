# 4. Análisis de variables y relaciones:


```python
# Gráfico de barras para las ventas globales por plataforma
df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).plot(kind='bar', figsize=(10,6))
plt.title('Ventas Globales por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (millones de unidades)')
plt.show()
```

En el gráfico aqui presente se muestra cuantos videojuegos fueron vendidos en todas las plataformas hasta el año 2016. Dejando en evidencia que las plataformas que dominaron el mercado siempre pertenecieron a las mismas tres (3) compañias: Sony, Microsoft y Nintendo; con sus respectivas Playstation3, Xbox360 y la Nintendo DS y Wii. Esto, al compararlo con un breve análisis económico y de costos de la época tiene mucho sentido ya que eran consolas relativamente baratas de adquirir (no mas de 300U$D la más cara), y los videojuegos al ser en un formato CD/DVD tambbien tenian un costo menor que los BluRay de hoy en día. Esto explica brevemente el boom de los lanzamientos de videojuegos asi como el nacimiento de muchos estudios desarrolladores.

```python
# Gráfico de dispersión para las ventas globales y calificación de usuarios
sns.scatterplot(x='User_Score', y='Global_Sales', data=df)
plt.title('Ventas Globales vs. Calificación de Usuarios')
plt.xlabel('Calificación de Usuarios')
plt.ylabel('Ventas Globales (millones de unidades)')
plt.show()
```

Este gráfico nos permite comprobar que a mayor calificación otorgada a los títulos por los usuarios, mayor sera el nivel de ventas de ese videojuego en el mercado global. Probando asi que los usuarios muchas veces compran mirando solo las reviews o de otros usuarios ya que se consideran mas 'honestas'.

```python
# Gráfico de dispersión para las ventas globales y calificación de críticos
sns.scatterplot(x='Critic_Score', y='Global_Sales', data=df)
plt.title('Ventas Globales vs. Calificación de Críticos')
plt.xlabel('Calificación de Críticos')
plt.ylabel('Ventas Globales (millones de unidades)')
plt.show()
```

Un poco siguiendo la misma explicación anterior, a mayor puntaje de critica mayor será la veta de ese videojuego en el mercado.

```python
# Gráfico de dispersión para las ventas globales y año de lanzamiento
sns.scatterplot(x='Year_of_Release', y='Global_Sales', data=df)
plt.title('Ventas Globales vs. Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Ventas Globales (millones de unidades)')
plt.show()
```

En este gráfico podemos ver que la 'época dorada' de los videojuegos fue entre el año 2004-2012, ya que fue el périodo en donde mas videojuegos se lanzaron al mercado, obteniendo así un gran volumen de ventas.

```python
# Gráfico de barras para las ventas globales por género
df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).plot(kind='bar', figsize=(10,6))
plt.title('Ventas Globales por Género')
plt.xlabel('Género')
plt.ylabel('Ventas Globales (millones de unidades)')
plt.show()
```

En este gráfico de barras se muestra como responde el mercado a los distintos géneros y subgéneros de videojuegos. Mostrando así que el género Acción es el que mayor número de titulos ofrecía al mercado (casi un cuarto del mercado global de videojuegos es representado por este género), seguido por los videojuegos de deportes (muy populares a nivel mundial) y dejando en último puesto a los juegos tipo Puzzle. Esta tendencía hoy en día se mantiene, pero la popularidad de los RPG (Role play game) como de los Simuladores ha crecido mucho más.

#  6. Modelo de clusterización KNN

El algoritmo de k vecinos más cercanos, también conocido como KNN o k-NN, es un clasificador de aprendizaje supervisado no paramétrico, que utiliza la proximidad para hacer clasificaciones o predicciones sobre la agrupación de un punto de datos individual.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

Preparar los datos:

En este caso, vamos a utilizar las variables "Global_Sales", "Critic_Score" y "User_Score" para predecir la variable "Platform". Primero, eliminamos las filas con valores nulos en estas variables y convertimos la variable "Platform" a una variable categórica.

```python
df = df.dropna(subset=["Global_Sales", "Critic_Score", "User_Score"])
df["Platform"] = df["Platform"].astype("category")
```

```python
#Dividimos el conjunto de entrenamiento y de prueba:

X = df[["Global_Sales", "Critic_Score", "User_Score"]]
y = df["Platform"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#Entrenamiento del modelo:
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

```python
#Probando el modelo:
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
print("\nInforme de Clasificación:")
print(classification_report(y_test, y_pred))
```

Los resultados obtenidos del modelo KNN aplicado al dataset de ventas de videojuegos muestran que la precisión del modelo es del 73.59%, lo que significa que alrededor del 73.59% de las clasificaciones del modelo son correctas. **Análisis del informe de clasificación:**

El informe de clasificación proporciona más detalles sobre la capacidad del modelo KNN para clasificar correctamente las diferentes clases en el conjunto de datos. Este informe muestra varias métricas de evaluación del modelo, incluyendo precisión, recall y f1-score para cada clase, así como la precisión media y el promedio ponderado de todas las métricas.

La precisión se refiere a la proporción de verdaderos positivos (TP) entre la suma de verdaderos positivos y falsos positivos (FP). Es decir, la precisión mide la capacidad del modelo para identificar correctamente una clase en relación con otras clases. Un alto valor de precisión indica que el modelo tiene una baja tasa de falsos positivos y es capaz de identificar correctamente la clase.

El recall, también conocido como sensibilidad, mide la proporción de verdaderos positivos entre la suma de verdaderos positivos y falsos negativos (FN). Es decir, el recall mide la capacidad del modelo para detectar correctamente una clase en relación con todas las instancias de esa clase. Un alto valor de recall indica que el modelo tiene una baja tasa de falsos negativos y es capaz de detectar correctamente la clase.

El f1-score es una medida combinada de precisión y recall que proporciona una puntuación equilibrada entre ambas métricas. El f1-score se calcula como la media armónica de precisión y recall, y varía entre 0 y 1. Un valor de f1-score cercano a 1 indica que el modelo tiene un buen equilibrio entre precisión y recall.

En el informe de clasificación, se puede observar que el modelo KNN tiene una precisión promedio del 73.58% y un recall promedio del 69.94%. Esto indica que el modelo es capaz de identificar correctamente la mayoría de las instancias de las dos clases, aunque tiene una tendencia a ser más preciso para la clase menos popular.

#  9. Métricas de validación

**Validación cruzada (Cross-validation)**: La validación cruzada es un método para evaluar la capacidad de generalización de un modelo estadístico utilizando un conjunto de datos limitado. En este método, el conjunto de datos se divide en k subconjuntos de igual tamaño, y se realiza el ajuste y la evaluación del modelo en k iteraciones, donde en cada iteración se utiliza un subconjunto diferente como conjunto de validación y los demás subconjuntos se utilizan como conjunto de entrenamiento. El resultado final es la media del desempeño del modelo en cada iteración. A continuación se muestra un ejemplo de cómo realizar la validación cruzada en el modelo de regresión lineal:

```python
from sklearn.model_selection import cross_val_score

# Calcular la puntuación de validación cruzada para el modelo
scores = cross_val_score(model, X, y, cv=5)
print('Puntuaciones de validación cruzada:', scores)

# Calcular la media y la desviación estándar de las puntuaciones
print('Media de puntuaciones:', scores.mean())
print('Desviación estándar de puntuaciones:', scores.std())
```

La función devuelve una matriz de puntuaciones de tamaño igual al número de subconjuntos, donde cada puntuación representa la precisión del modelo en un subconjunto específico. Además, se calcula la media y la desviación estándar de las puntuaciones para obtener una idea de la estabilidad del modelo.

**Tambien aplicaremos la División en conjunto de entrenamiento y conjunto de prueba**: La división en conjunto de entrenamiento y conjunto de prueba es un método para evaluar la capacidad de generalización de un modelo estadístico mediante la separación del conjunto de datos en dos partes: una parte se utiliza para entrenar el modelo (conjunto de entrenamiento), y la otra parte se utiliza para evaluar la precisión del modelo (conjunto de prueba). En este método, el conjunto de entrenamiento se utiliza para ajustar el modelo, y el conjunto de prueba se utiliza para evaluar la precisión del modelo en datos no vistos durante el entrenamiento. A continuación se muestra un ejemplo de cómo realizar la división en conjunto de entrenamiento y conjunto de prueba en el modelo de regresión lineal:

```python
from sklearn.model_selection import train_test_split

# Dividir el conjunto de datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Entrenar el modelo con el conjunto de entrenamiento
model.fit(X_train, y_train)

# Evaluar la precisión del modelo con el conjunto de prueba
score = model.score(X_test, y_test)
print('Precisión del modelo:', score)
```

Un modelo con un score de 0.9999870734418053 indica que el modelo es muy preciso en la predicción de los datos de entrenamiento. Sin embargo, esto no es necesariamente una indicación de que el modelo tendrá una capacidad de generalización similar en datos no vistos. En cuanto a la selección del modelo, un score alto en el conjunto de entrenamiento no siempre significa que el modelo es el mejor para los datos. Es importante considerar otros factores, como la complejidad del modelo, la interpretabilidad, la eficiencia y la escalabilidad. En resumen, aunque un score alto en el conjunto de entrenamiento puede ser alentador, es importante evaluar la capacidad de generalización del modelo y considerar otros factores en la selección del modelo.

**Tecnica de Silhouette**: para evaluar la calidad del clustering generado por el modelo K-means. Esta técnica mide cómo de bien se ajusta cada punto de datos a su propio clúster en comparación con otros clústeres. Cuanto mayor sea el valor de Silhouette, mejor será la separación entre los clústeres.

```python
from sklearn.metrics import silhouette_score

# Calcular el valor de Silhouette para diferentes valores de k
silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df[relevant_features.index])
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(df[relevant_features.index], labels))

# Visualizar los resultados de Silhouette analysis
plt.plot(range(2, 10), silhouette_scores)
plt.title('Silhouette Analysis')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

El valor de Silhouette es un indicador adicional de la calidad del clustering generado por el modelo K-means. En este caso, Silhouette analysis sugiere que dividir los datos en dos clústeres es la mejor opción, lo que no coincide con el resultado del modelo K-means anterior el cual dividimos en 3 clusters. La combinación de diferentes técnicas de validación, como la división en conjuntos de entrenamiento y prueba y Silhouette analysis, puede proporcionar una evaluación más completa de la calidad del modelo y su capacidad de generalización.
