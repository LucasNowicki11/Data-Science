En el siguiente notebook se agregaran lineas de código para darle una visualización a las respuestas de las preguntas de interés de la entrega anterior, que junto con los insights obtenidos al principio del curso serviran de base para la explicación de las respuestas.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/Lucas/Desktop/Proyecto_Data_Science/Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
#Pregunta 1: ¿Qué factores influyen en la popularidad y las ventas de los videojuegos?
# Histograma de las ventas totales en todo el mundo
plt.hist(df['Global_Sales'], bins=20, edgecolor='black')
plt.xlabel('Ventas totales en todo el mundo (millones)')
plt.ylabel('Frecuencia')
plt.title('Distribución de las ventas totales en todo el mundo')
plt.show()
```

```python
# Gráfico de barras para las ventas globales por plataforma
df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).plot(kind='bar', figsize=(10,6))
plt.title('Ventas Globales por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (millones de unidades)')
plt.show()
```

En cuanto al análisis de consolas, GameTechno descubrió que las consolas PlayStation y Xbox eran las más populares, y que los juegos exclusivos para estas consolas tenían una mayor demanda. También notaron que los juegos en línea estaban creciendo en popularidad, lo que sugiere que pueden tener un impacto positivo en las ventas de videojuegos y lo cual sería una buena oportunidad de desarrollo.
Las consolas como la PS3, la PS4, la XBOX 360 fueron las que mejores números de ventas tuvieron en un tiempo mas sostenido que las demas, las cuales tuvieron una popularidad mucho mas acortada.

```python
#Pregunta 2: ¿Existe una relación entre el género de los videojuegos y los gustos de los videojuegos?


# Gráfico de barras apiladas para mostrar los géneros de videojuegos y sus ventas totales
genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_sales.index, y=genre_sales.values, palette='viridis')
plt.xlabel('Género de videojuegos')
plt.ylabel('Ventas totales en todo el mundo (millones)')
plt.title('Ventas totales por género de videojuegos')
plt.xticks(rotation=45)
plt.show()
```

GameTechno decidió investigar utilizando un conjunto de datos extraidos de Kaggle que tenía información sobre ventas de videojuegos de todo el mundo. Al analizar los datos, descubrieron que los videojuegos de acción eran los más populares en América y Europa, mientras que los juegos de estrategia eran los más populares en Asia. Además, los videojuegos de deportes eran populares en todo el mundo, pero especialmente en América del Norte.

A raíz de esto, la empresa se pregunta si ¿Existe una relación entre la popularidad de las plataformas de videojuegos y los géneros más vendidos?

```python

# Tabla de pivote que muestre las ventas totales por plataforma y género
pivot_table = df.pivot_table(index='Platform', columns='Genre', values='Global_Sales', aggfunc='sum')

# Heatmap que muestre las ventas totales por plataforma y género
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.1f', cbar=True)
plt.xlabel('Género')
plt.ylabel('Plataforma')
plt.title('Ventas totales por plataforma y género')
plt.show()
```

A partir de este heatmap se pueden identificar claramente las combinaciones de género y plataforma más exitosas para realizar desarrollo y lanzamiento de videojuegos.

```python
#Pregunta 3: ¿Existe una correlación entre la calificación de los videojuegos y sus ventas?

# Gráfico de dispersión entre las calificaciones y las ventas totales en todo el mundo
plt.scatter(df['Critic_Score'], df['Global_Sales'], alpha=0.5)
plt.xlabel('Calificación de los críticos')
plt.ylabel('Ventas totales en todo el mundo (millones)')
plt.title('Relación entre la calificación de los videojuegos y sus ventas')
plt.show()
```

```python
# Calcular el coeficiente de correlación entre las calificaciones y las ventas
correlation = df['Critic_Score'].corr(df['Global_Sales'])
print("Coeficiente de correlación: ", correlation)
```

Con respecto a la correlación entre la calificación de los videojuegos y las ventas, GameTechno descubrió que había una correlación positiva. Los juegos con calificaciones más altas tenían más ventas, lo que sugiere que los jugadores confían en las opiniones de los expertos y de la comunidad.


```python
# Pregunta 4: ¿Es más rentable el lanzamiento exclusivo de videojuegos en Norteamérica o en otra región?

# Gráfico de barras que muestre las ventas en Norteamérica y otras regiones
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='muted')
plt.xlabel('Región')
plt.ylabel('Ventas totales en todo el mundo (millones)')
plt.title('Ventas totales por región')
plt.show()
```

Por último, en cuanto a la pregunta sobre el lanzamiento exclusivo de videojuegos, GameTechno llegó a la conclusión de que el mercado Norteamericano representaba el mayor porcentaje de las ventas a nivel global en videojuegos. Sin embargo, también encontraron que lanzar videojuegos exclusivamente en Norteamérica no necesariamente era más rentable que lanzarlos en otras regiones. En cambio, descubrieron que era más importante adaptarse a las preferencias y hábitos de compra de los jugadores en cada región.

Con toda esta información, la empresa se pregunta ¿qué factores influyen en la popularidad y la venta de los videojuegos?

```python
# Gráfico de boxplot para comparar las ventas totales en todo el mundo según el género
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=df)
plt.xlabel('Género')
plt.ylabel('Ventas totales en todo el mundo (millones)')
plt.title('Comparación de ventas totales por género')
plt.show()
```

```python
# Gráfico de boxplot para comparar las ventas totales en todo el mundo según la plataforma
plt.figure(figsize=(10, 6))
sns.boxplot(x='Platform', y='Global_Sales', data=df)
plt.xlabel('Plataforma')
plt.ylabel('Ventas totales en todo el mundo (millones)')
plt.title('Comparación de ventas totales por plataforma')
plt.xticks(rotation=90)
plt.show()
```

Comparación de ventas totales por género:

Los géneros de "Plataformas" y "Shooter" tienden a tener ventas más altas en comparación con otros géneros.
Los géneros de "Estrategia" y "Aventura" tienden a tener ventas más bajas en comparación con otros géneros.
Existen diferencias significativas en las ventas totales entre los diferentes géneros de videojuegos.

Comparación de ventas totales por plataforma:

Las plataformas "PS2" y "X360" tienden a tener ventas más altas en comparación con otras plataformas.
Las plataformas "SNES" y "GB" tienden a tener ventas más bajas en comparación con otras plataformas.
